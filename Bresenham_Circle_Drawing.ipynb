```python
###Bresenham Circle Drawing###
import matplotlib.pyplot as plt
print("Enter Xc: ")
xc=int(input())
print("Enter Yc: ")
yc =int(input())
print("Enter radius: ")
r=int(input())
X=[]
Y=[]
x=0
y=r
d=3-2*r;
while x<=y:
  X.extend([x+xc, y+xc, x+xc, y+xc, -x+xc, -y+xc,-x+xc, -y+xc])
  Y.extend([y+yc, x+yc, -y+yc, -x+yc, -y+yc, -x+yc, y+yc, x+yc])
  if d<0:
    d=d+4*x+6;
  else:
    y=y-1;
    d=d+4*(x-y)+10;
  x=x+1;
plt.title("Bresenham's Circle Drawing Algorithm")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.scatter(X,Y)
```
