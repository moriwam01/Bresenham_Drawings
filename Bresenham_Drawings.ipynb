```python
####Bresenham Line Drawing####
import matplotlib.pyplot as plt
print("Enter positive value of x1:")
x1 = int(input())
print("Enter positive value of y1:")
y1 = int(input())
print("Enter positive value of x2:")
x2 = int(input())
print("Enter positive value of x4:")
y2 = int(input())
dx = abs(x2-x1)
dy = abs(y2-y1)
dS = 2*dy
dT = 2*(dy-dx)
x = x1
y = y1
xcoordinates = [x]
ycoordinates = [y]
d = 2*dy-dx
while x < x2:
  x = x+1
  if d < 0:
      d = d+dS
  else:
      y = y+1
      d = d+dT
  xcoordinates.append(x)
  ycoordinates.append(y)
plt.title("Bresenham's Line Drawing Algorithm")
plt.plot(xcoordinates, ycoordinates)
```

```python
###Bresenham Circle Drawing###
import matplotlib.pyplot as plt
print("Enter Xc: ")
xc=int(input())
print("Enter Yc: ")
yc =int(input())
print("Enter radius: ")
r=int(input())
X=[]
Y=[]
x=0
y=r
d=3-2*r;
while x<=y:
  X.extend([x+xc, y+xc, x+xc, y+xc, -x+xc, -y+xc,-x+xc, -y+xc])
  Y.extend([y+yc, x+yc, -y+yc, -x+yc, -y+yc, -x+yc, y+yc, x+yc])
  if d<0:
    d=d+4*x+6;
  else:
    y=y-1;
    d=d+4*(x-y)+10;
  x=x+1;
plt.title("Bresenham's Circle Drawing Algorithm")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.scatter(X,Y)
```

```python
###Bresenham Midpoint Circle Drawing###
import matplotlib.pyplot as plt
print("Enter Xc: ")
xc= int(input())
print("Enter Yc: ")
yc = int(input())
print("Enter radius: ")
r = int(input())
X = []
Y = []
x = 0
y = r
d = 1 - r;
while x<=y:
  X.extend([ x + xc, y + xc, x + xc, y + xc, -x + xc, -y + xc,-x + xc, -y + xc])
  Y.extend([y + yc, x + yc, -y + yc, -x + yc, -y + yc, -x + yc, y + yc, x + yc])
  if d < 0:
   d = d + 2*x + 3;
  else:
    y = y -1;
    d = d + 2 * (x - y) + 5;
  x = x + 1;
plt.title("Midpoint Circle Drawing Algorithm")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.scatter(X,Y)
```
