```python
###Bresenham Midpoint Circle Drawing###
import matplotlib.pyplot as plt
print("Enter Xc: ")
xc= int(input())
print("Enter Yc: ")
yc = int(input())
print("Enter radius: ")
r = int(input())
X = []
Y = []
x = 0
y = r
d = 1 - r;
while x<=y:
  X.extend([ x + xc, y + xc, x + xc, y + xc, -x + xc, -y + xc,-x + xc, -y + xc])
  Y.extend([y + yc, x + yc, -y + yc, -x + yc, -y + yc, -x + yc, y + yc, x + yc])
  if d < 0:
   d = d + 2*x + 3;
  else:
    y = y -1;
    d = d + 2 * (x - y) + 5;
  x = x + 1;
plt.title("Midpoint Circle Drawing Algorithm")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.scatter(X,Y)
```
