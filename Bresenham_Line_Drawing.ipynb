```python
import matplotlib.pyplot as plt
print("Enter positive value of x1:")
x1 = int(input())
print("Enter positive value of y1:")
y1 = int(input())
print("Enter positive value of x2:")
x2 = int(input())
print("Enter positive value of x4:")
y2 = int(input())
dx = abs(x2-x1)
dy = abs(y2-y1)
dS = 2*dy
dT = 2*(dy-dx)
x = x1
y = y1
xcoordinates = [x]
ycoordinates = [y]
d = 2*dy-dx
while x < x2:
  x = x+1
  if d < 0:
      d = d+dS
  else:
      y = y+1
      d = d+dT
  xcoordinates.append(x)
  ycoordinates.append(y)

plt.plot(xcoordinates, ycoordinates)
```
